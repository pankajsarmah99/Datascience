```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = [1,2,3,4,5] , y = [3,4,5,6,7], mode = 'lines'))
fig.show()
#mode chage here 
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = [1,2,3,4,5] , y = [3,4,5,6,7]))
fig.show()
```

```python
from bokeh.plotting import figure, output_file , show
from bokeh.sampledata.iris import flowers

output_file('test.html')

p = figure(title = "test flower")
p.xaxis.axis_label = "petal length"
p.yaxis.axis_label = "petal width"
p.circle(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
from bokeh.plotting import figure, output_file , show
from bokeh.sampledata.iris import flowers

output_file('line.html')

p = figure(title = "line graph")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.line(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python
x = [2,3,4,5,6]
y = [3,4,5,6,7]
output_file("line.html")
p = figure(title = "line plot")
p.line(x,y)
show(p)
```

```python
x = [2,3,4,5,6]
y = [3,4,5,6,7]
output_file("line.html")
p = figure(title = "line plot")
p.scatter(x,y , fill_color = 'red' , legend_label = 'red points' , size = 20)
show(p)
```
